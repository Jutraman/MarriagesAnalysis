```python
import pandas as pd
import numpy as np
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.commons.utils import JsCode
from pyecharts.globals import ThemeType
import pyecharts.options as opts
from pyecharts.charts import Line
```

```python
data1 = pd.read_excel('./marriagesworkbook2017forpublication.xlsx', sheet_name='Table 1')
data_new = pd.DataFrame(np.array(data1.iloc[7:28, [0,3,5,10]]).T, index = ['year', 'civil_cere', 'reli_cere', 'same_sex'])
data_new = data_new.T
```

```python
year1 = []
list1 = []
list2 = []
list3 = []
# for i in np.array(data_new)[::-1]:
for i in np.array(data_new)[::-1]:
    tmp_dict1 = dict()
    tmp_dict2 = dict()
    tmp_dict3 = dict()
    year1.append(int(str(i[0])[:4]))
    a = i[1]
    b = i[2]
    c = i[3]
    if i[3] == 'z':
        c = 0
    sum_ = a + b + c
    tmp_dict1 = dict([['value', a], ['percent', a/sum_]])
    tmp_dict2 = dict([['value', b], ['percent', b/sum_]])
    tmp_dict3 = dict([['value', c], ['percent', c/sum_]])
    list1.append(tmp_dict1)
    list2.append(tmp_dict2)
    list3.append(tmp_dict3)
```

```python
data2 = pd.read_excel('./marriagesworkbook2017forpublication.xlsx', sheet_name='Table 5')
data_new = pd.DataFrame(np.array(data2.iloc[7:-20, [0,1,5]]).T, index = ['year', 'men marriage age','women marriage age'])
data_new = data_new.T
data_new

year3 = []
for i in np.array(data_new)[::-1]:
    year3.append(int(str(i[0])[:4]))

list5 = list(data_new['men marriage age'])[::-1]
list8 = list(data_new['women marriage age'])[::-1]
```

```python
data2 = pd.read_excel('./marriagesworkbook2017forpublication.xlsx', sheet_name='Table 8')
data_new = pd.DataFrame(np.array(data2.iloc[9:-20, :4]).T, index = ['year', 'marriage rate', 'man marriage rate', 'woman marriage rate'])
data_new = data_new.T
data_new

year4 = []
for i in np.array(data_new)[::-1]:
    year4.append(int(str(i[0])[:4]))

list6 = list(data_new['man marriage rate'])[::-1]
list7 = list(data_new['woman marriage rate'])[::-1]
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Grid, Line, Liquid, Page, Pie
from pyecharts.commons.utils import JsCode
from pyecharts.components import Table
from pyecharts.faker import Faker

def bar_stack() -> Bar:
    c = (
        Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
        .add_xaxis(year1)
        .add_yaxis("Civil ceremonies", list1, stack="stack1", category_gap="50%")
        .add_yaxis("Religious ceremonies", list2, stack="stack1", category_gap="50%")
        .add_yaxis("All marriages to same-sex couples", list3, stack="stack1", category_gap="50%")

        .set_series_opts(
            label_opts=opts.LabelOpts(
                position="right",
                formatter=JsCode(
                    "function(x){return Number(x.data.percent * 100).toFixed() + '%';}"
                ),
            )
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Bar_ceremony"),
        )
    )
    return c
    
def bar_datazoom_slider() -> Bar:
    c = (
        Bar()
        .add_xaxis(year2)
        .add_yaxis("All_marriages", list4)
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Bar_all_marriages"),
            datazoom_opts=[opts.DataZoomOpts()],
        )
    )
    return c

def line1() -> Line:
    c = (
        Line()
        .set_global_opts(
            tooltip_opts=opts.TooltipOpts(is_show=True),
            xaxis_opts=opts.AxisOpts(type_="category"),
            yaxis_opts=opts.AxisOpts(
                type_="value",
                axistick_opts=opts.AxisTickOpts(is_show=True),
                splitline_opts=opts.SplitLineOpts(is_show=True),
            ),
        )
        .add_xaxis(xaxis_data=[str(i) for i in year3])
        .add_yaxis(
            series_name="Median age of men",
            y_axis=list5,
            symbol="emptyCircle",
            is_symbol_show=True,
            label_opts=opts.LabelOpts(is_show=False),
        )
         .add_yaxis(
            series_name="Median age of women",
            y_axis=list8,
            symbol="emptyCircle",
            is_symbol_show=True,
            label_opts=opts.LabelOpts(is_show=False),
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Marriage_age_change_line"),
            datazoom_opts=[opts.DataZoomOpts()],
            tooltip_opts=opts.TooltipOpts(trigger="axis"),
            xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),
            yaxis_opts=opts.AxisOpts(
                type_="value",
                min_=20,
                axistick_opts=opts.AxisTickOpts(is_show=True),
                splitline_opts=opts.SplitLineOpts(is_show=True),
            ),
        )
    )
    return c


def line2() -> Line:
    c = (
        Line()
        .set_global_opts(
            tooltip_opts=opts.TooltipOpts(is_show=True),
            xaxis_opts=opts.AxisOpts(type_="category"),
            yaxis_opts=opts.AxisOpts(
                type_="value",
                min_=20,
                axistick_opts=opts.AxisTickOpts(is_show=True),
                splitline_opts=opts.SplitLineOpts(is_show=True),
            ),
            
        )
        .add_xaxis(xaxis_data=[str(i) for i in year4])
        .add_yaxis(
            series_name="Men marriage rates",
            y_axis=list6,
            symbol="emptyCircle",
            is_symbol_show=True,
            label_opts=opts.LabelOpts(is_show=False),
            
        )
        .add_yaxis(
            series_name="Women marriage rates",
            y_axis=list7,
            symbol="emptyCircle",
            is_symbol_show=True,
            label_opts=opts.LabelOpts(is_show=False),
            
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Marriage_rate_change_line"),
            datazoom_opts=[opts.DataZoomOpts()],
            yaxis_opts=opts.AxisOpts(
                type_="value",
                axistick_opts=opts.AxisTickOpts(is_show=True),
                splitline_opts=opts.SplitLineOpts(is_show=True),
            ),
        )
    )
    return c


def page_default_layout():
    page = Page()
    page.add(
        bar_datazoom_slider(),
        line2(),
        bar_stack(),
        line1(),
        )
    page.render("COUSEWORK 2.html")


if __name__ == "__main__":
    page_default_layout()
```
